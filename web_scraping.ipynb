```python
import requests
import os
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
```

```python
years = list(range(2003, 2024)) #years that will be scraped, and used to train data
```

```python
url_start = "https://www.baseball-reference.com/awards/awards_{}.shtml" #Cy Young award voting history
```

```python
driver = webdriver.Edge() #user may need to add their own path, considering I run jupyter on a local host
```

```python
#loads and saves the Cy Young webpages
for year in years:
    url = url_start.format(year)

    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)

    html = driver.page_source
    with open("cyyoung/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(html)
```

```python
#Parses html and loads AL Cy Young winners to pandas data frames for respective years
dfs = []
for year in years:
    with open("cyyoung/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
    soup = BeautifulSoup(page, "html.parser")
    alcy_table = soup.find(id="AL_CYA_voting")
    alcy = pd.read_html(str(alcy_table))[0]
    alcy["Year"] = year
    dfs.append(alcy)
```

```python
alcys = pd.concat(dfs) #Combines all AL Cy Young voting history to a single data frame
```

```python
#Parses html and loads NL Cy Young winners to pandas data frames for respective years
dfs = []
for year in years:
    with open("cyyoung/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
    soup = BeautifulSoup(page, "html.parser")
    nlcy_table = soup.find(id="NL_CYA_voting")
    nlcy = pd.read_html(str(nlcy_table))[0]
    nlcy["Year"] = year
    dfs.append(nlcy)
```

```python
nlcys = pd.concat(dfs) #Combines all AL Cy Young voting history to a single data frame
```

```python
alcys.to_csv("alcys.csv") #Saves AL Cy Young canidates to CSV
nlcys.to_csv("nlcys.csv") #Saves NL Cy Young canidates to CSV
```

```python
player_stats_url = "https://www.baseball-reference.com/leagues/majors/{}-standard-pitching.shtml"
```

```python
#loads and saves all pitching stats webpages
for year in years:
    url = player_stats_url.format(year)

    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)

    html = driver.page_source
    with open("player/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(html)
```

```python
#Parses html and loads all pitching stats to pandas data frames for respective years
dfs = []
for year in years:
    with open("player/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
        soup = BeautifulSoup(page, "html.parser")
        soup.find("tr", class_="thead").decompose() #Removes annoying headers that appear every few rows
        soup.find("tr", class_="league_average_table italic_text").decompose()
        player_table = soup.find(id="players_standard_pitching")
        player = pd.read_html(str(player_table))[0]
        player["Year"] = year
        dfs.append(player)
```

```python
players = pd.concat(dfs) #Combines all pitching stats history to a single data frame
```

```python
players
```

```python
players.to_csv("players.csv")
```

```python
team_stats_url = "https://www.baseball-reference.com/leagues/majors/2023-standings.shtml"
```

```python
#loads and saves all team stats webpages
for year in years:
    url = team_stats_url.format(year)

    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)

    html = driver.page_source
    with open("team/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(html)
```

```python
#Parses html and loads all team stats to pandas data frames for respective years
dfs = []
for year in years:
    with open("team/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
        soup = BeautifulSoup(page, "html.parser")
        soup.find("tr", class_="thead").decompose() #Removes annoying headers that appear every few rows
        soup.find("tr", class_="league_average_table").decompose() #Removes league average row
        team_table = soup.find(id="expanded_standings_overall")
        team = pd.read_html(str(team_table))[0]
        team["Year"] = year
        dfs.append(team)
```

```python
teams = pd.concat(dfs) #Combines all team stats history to a single data frame
```

```python
teams.to_csv("teams.csv")
```
