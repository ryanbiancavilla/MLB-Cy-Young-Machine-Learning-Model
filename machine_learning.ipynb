```python
#Gives machine points for correctly predicting the top 10 AL Cy Young vote receivers, and penalizes it for missing
def find_alcy(al_combination):
    al_actual = al_combination.sort_values("Share", ascending=False).head(10)
    al_predicted = al_combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index, row in al_predicted.iterrows():
        if row["Name"] in al_actual["Name"].values:
            found += 1
            ps.append(found/seen)
        seen += 1
    return sum(ps) / len(ps)
```

```python
find_alcy(al_combination)
```

```python
#Gives machine points for correctly predicting the top 10 AL Cy Young vote receivers, and penalizes it for missing
def find_nlcy(nl_combination):
    nl_actual = nl_combination.sort_values("Share", ascending=False).head(10)
    nl_predicted = nl_combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index, row in nl_predicted.iterrows():
        if row["Name"] in nl_actual["Name"].values:
            found += 1
            ps.append(found/seen)
        seen += 1
    return sum(ps) / len(ps)
```

```python
al_aps = []
al_all_predictions = []

for year in years[5:]:
    al_train = al_stats[al_stats["Year"] < year]
    
    #Remove rows with infinite values in the ERA column
    al_train = al_train.replace([np.inf, -np.inf], np.nan)
    al_train = al_train.dropna(subset=['ERA'])
    
    al_test = al_stats[al_stats["Year"] == year]
    
    #Remove rows with infinite values in the ERA column from the test set
    al_test.loc[:, 'ERA'] = al_test['ERA'].replace([np.inf, -np.inf], np.nan)
    al_test = al_test.dropna(subset=['ERA'])
    
    reg.fit(al_train[predictors], al_train["Share"])
    
    al_predictions = reg.predict(al_test[predictors])
    al_predictions = pd.DataFrame(al_predictions, columns=["predictions"], index=al_test.index)
    al_combination = pd.concat([al_test[["Name", "Share"]], al_predictions], axis=1)
    al_all_predictions.append(al_combination)
    al_aps.append(find_alcy(al_combination))
```

```python
nl_aps = []
nl_all_predictions = []

for year in years[5:]:
    nl_train = nl_stats[nl_stats["Year"] < year]
    
    #Remove rows with infinite values in the ERA column
    nl_train = nl_train.replace([np.inf, -np.inf], np.nan)
    nl_train = nl_train.dropna(subset=['ERA'])
    
    nl_test = nl_stats[nl_stats["Year"] == year]
    
    #Remove rows with infinite values in the ERA column from the test set
    nl_test.loc[:, 'ERA'] = nl_test['ERA'].replace([np.inf, -np.inf], np.nan)
    nl_test = nl_test.dropna(subset=['ERA'])
    
    reg.fit(nl_train[predictors], nl_train["Share"])
    
    nl_predictions = reg.predict(nl_test[predictors])
    nl_predictions = pd.DataFrame(nl_predictions, columns=["predictions"], index=nl_test.index)
    nl_combination = pd.concat([nl_test[["Name", "Share"]], nl_predictions], axis=1)
    nl_all_predictions.append(nl_combination)
    nl_aps.append(find_nlcy(nl_combination))
```

```python
def al_add_ranks(al_combination):
    al_combination = al_combination.sort_values("Share", ascending=False)
    al_combination["Rk"] = list(range(1, al_combination.shape[0]+1))
    al_combination = al_combination.sort_values("predictions", ascending=False)
    al_combination["Predicted_Rk"] = list(range(1, al_combination.shape[0]+1))
    al_combination["Diff"] = al_combination["Rk"] - al_combination["Predicted_Rk"]
    return al_combination
```

```python
al_ranking = al_add_ranks(al_all_predictions[1])
al_ranking[al_ranking["Rk"] < 11].sort_values("Diff", ascending=False)
```

```python
def nl_add_ranks(nl_combination):
    nl_combination = nl_combination.sort_values("Share", ascending=False)
    nl_combination["Rk"] = list(range(1, nl_combination.shape[0]+1))
    nl_combination = nl_combination.sort_values("predictions", ascending=False)
    nl_combination["Predicted_Rk"] = list(range(1, nl_combination.shape[0]+1))
    nl_combination["Diff"] = nl_combination["Rk"] - nl_combination["Predicted_Rk"]
    return nl_combination
```

```python
def al_backtest(al_stats, model, year, predictors):
    al_aps = []
    al_all_predictions = []
    for year in years[5:]:
        al_train = al_stats[al_stats["Year"] < year]
        #Remove rows with infinite values in the ERA column
        al_train = al_train.replace([np.inf, -np.inf], np.nan)
        al_train = al_train.dropna(subset=['ERA'])
        al_test = al_stats[al_stats["Year"] == year]
        #Remove rows with infinite values in the ERA column from the test set
        al_test.loc[:, 'ERA'] = al_test['ERA'].replace([np.inf, -np.inf], np.nan)
        al_test = al_test.dropna(subset=['ERA'])
        model.fit(al_train[predictors], al_train["Share"])
        al_predictions = reg.predict(al_test[predictors])
        al_predictions = pd.DataFrame(al_predictions, columns=["predictions"], index=al_test.index)
        al_combination = pd.concat([al_test[["Name", "Share"]], al_predictions], axis=1)
        al_combination = al_add_ranks(al_combination)
        al_all_predictions.append(al_combination)
        al_aps.append(find_alcy(al_combination))
    return sum(al_aps)/len(al_aps), al_aps, pd.concat(al_all_predictions)
```

```python
al_mean_ap, al_aps, al_all_predictions = al_backtest(al_stats, reg, years[5:], predictors)
```

```python
def nl_backtest(nl_stats, model, year, predictors):
    nl_aps = []
    nl_all_predictions = []
    for year in years[5:]:
        nl_train = nl_stats[nl_stats["Year"] < year]
        #Remove rows with infinite values in the ERA column
        nl_train = nl_train.replace([np.inf, -np.inf], np.nan)
        nl_train = nl_train.dropna(subset=['ERA'])
        nl_test = nl_stats[nl_stats["Year"] == year]
        #Remove rows with infinite values in the ERA column from the test set
        nl_test.loc[:, 'ERA'] = nl_test['ERA'].replace([np.inf, -np.inf], np.nan)
        nl_test = nl_test.dropna(subset=['ERA'])
        model.fit(nl_train[predictors], nl_train["Share"])
        nl_predictions = reg.predict(nl_test[predictors])
        nl_predictions = pd.DataFrame(nl_predictions, columns=["predictions"], index=nl_test.index)
        nl_combination = pd.concat([nl_test[["Name", "Share"]], nl_predictions], axis=1)
        nl_combination = nl_add_ranks(nl_combination)
        nl_all_predictions.append(nl_combination)
        nl_aps.append(find_nlcy(nl_combination))
    return sum(nl_aps)/len(nl_aps), nl_aps, pd.concat(nl_all_predictions)
```
