```python
#Replaces TOT and Lg with the last team a traded player was on and returns the last team they were on and the Lg for that respective year
def single_team(df):
    if df.shape[0] == 1:
        return df
    else:
        #Identify the last team and league for each player-year combination
        tot_rows = df[df['Tm'] == 'TOT']
        
        if not tot_rows.empty:
            last_team_index = tot_rows.index[0]
            last_team = df.loc[last_team_index, 'Tm']
            last_league = df.loc[last_team_index, 'Lg']
            
            #Replace "TOT" with the last team and update "Lg"
            df.loc[df['Tm'] == 'TOT', 'Tm'] = last_team
            df.loc[df['Tm'] == last_team, 'Lg'] = last_league
        
        #Return only the last row of the group
        return df.iloc[[-1]]

#Apply the function to the dataframe
players = players.groupby(['Name', 'Year']).apply(single_team).reset_index(drop=True)
```

```python
#Maps teams' abbreviations to their full names in a hashmap from CSV file and indexes them
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```
